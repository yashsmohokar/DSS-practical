```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

line chart

```python
plt.plot(x,y)
plt.title("line chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.show()
```

bar chart


```python
plt.bar(x,y)
plt.title("bar chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.bar(x,y, color='red')
plt.show()

```

scatter plot

```python
a=(1,3,4,5,6,7,9,4)
b=(24,35,46,35,56,78,56,58)

plt.scatter(a,b)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

histogram

```python
h=[1,2,3,3,4,4,5,5,1,2,6,6,6,9,9]
```
